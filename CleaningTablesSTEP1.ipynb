```python
#import combined table to python

import pandas as pd
combined = pd.read_csv("./CombinedTableTBC.csv")
combined.head()
```

```python
#converting columns to datetime format

new_AppDate = combined['AppDate']
new_AppDate = pd.to_datetime(new_AppDate)
combined['AppDate']=new_AppDate

new_dob = combined['dateOfBirth']
new_dob = pd.to_datetime(new_dob)
combined['dateOfBirth']=new_dob

opened_date = combined['Opened']
opened_date = pd.to_datetime(opened_date)
combined['Opened']=opened_date

lp_date = combined['LastPayment']
lp_date = pd.to_datetime(lp_date)
combined['LastPayment']=lp_date

rc_date = combined['ReportedandCertified']
rc_date = pd.to_datetime(rc_date)
combined['ReportedandCertified']=rc_date

phs_date = combined['PaymentHistoryStart']
phs_date = pd.to_datetime(phs_date)
combined['PaymentHistoryStart']=phs_date

phe_date = combined['PaymentHistoryEnd']
phe_date = pd.to_datetime(phe_date)
combined['PaymentHistoryEnd']=phe_date
```

```python
# removing decimal points from PinCode and PermPinCode columns

pd.set_option('precision',0)
combined.head()
```

```python
# gender column

gender_info = combined['gender'].value_counts()
gender_info
```

```python
#calculating mode

gender_mode = combined['gender'].mode()
gender_mode
```

```python
#to deal with capitalisation issues

combined['gender'] = combined['gender'].str.upper()

#to deal with other cases

mapping = {'M':'MALE'}
combined['gender'] = combined['gender'].replace(mapping)



#replaceing missing values with mode

combined['gender'].fillna('MALE',inplace=True)

#drop row with outlier data

gender_values = ['MALE','FEMALE']
combined = combined[combined['gender'].isin(gender_values)] 
```

```python
#maritalStatus column

status = combined['maritalStatus'].value_counts()
status
```

```python
marital_mode = combined['maritalStatus'].mode()
marital_mode
```

```python
#to deal with capitalisation

combined['maritalStatus'] = combined['maritalStatus'].str.upper()

#other cases

marital_mapping = {'UNMARRIED':'SINGLE','SINGAL':'SINGLE','WID':'SINGLE','M':'MARRIED', 'MARIED':'MARRIED','MARRRIED':'MARRIED'}
combined['maritalStatus'] = combined['maritalStatus'].replace(marital_mapping)

#replace null values with mode

combined['maritalStatus'].fillna('MARRIED',inplace=True)

#replace other cases with mode

combined['maritalStatus'].replace({"SINGLE MARRIED": "MARRIED"}, inplace = True)

```
