```python
#import combined table to python

import pandas as pd
merge = pd.read_csv("D:\Internship\Dataset\FinalModeliteration\merge.csv")
merge.reset_index()
merge.columns = ['NUID','LoanNumber']
merge.drop(['NUID'],inplace=True,axis=1)
merge
```

```python
combined= combined.sort_values(by = ['UID','Opened'],ascending=False)
combined.drop(['LoanNumber'],inplace=True,axis=1)
combined
```

```python
combined = combined.reset_index(drop=True)
merge = merge.reset_index(drop=True)
```

```python
combined = pd.concat([combined, merge], axis=1)
combined.head(5)
```

```python
combined['LoanNumber'].head(5)
```

```python
combined['Opened'].head()
```

```python
# converting AppDate(text data) to dd/mm/yyyy format

new_AppDate = combined['AppDate']
new_AppDate = pd.to_datetime(new_AppDate)
combined['AppDate']=new_AppDate

#converting dateOfBirth(text data) to dd/mm/yyyy format

new_dob = combined['dateOfBirth']
new_dob = pd.to_datetime(new_dob)
combined['dateOfBirth']=new_dob


#converting opened column to datetime format

new_opened = combined['Opened']
new_opened = pd.to_datetime(new_opened)
combined['Opened']=new_opened
combined['Opened']
```

```python
#calculating age using dateOfBirth and AppDate and making a new column age

import numpy as np
combined['Age'] = (combined.Opened - combined.dateOfBirth)/np.timedelta64(1,'Y')

combined.head()
```

```python
loancount = previousloanmodel.groupby(['UID']).size().to_frame()
loancount = loancount.reset_index()
loancount.columns =['UID','number_of_loans']
loancount
```

```python
bal = previousloanmodel.groupby(['UID'])["CurrentBalance"].apply(lambda x : x.astype(int).sum()).to_frame()
bal = bal.reset_index()
bal.columns = ['UID','prev_bal']
bal
```

```python
deflist = previousloanmodel.groupby(['UID'])["loandefault"].apply(lambda x : x.astype(int).sum()).to_frame()
deflist = deflist.reset_index()
deflist.columns =['UID','defsum']
deflist
```

```python
loancount = pd.merge(loancount, deflist, on=['UID'])
loancount
```

```python
loancount['defaultdif'] = loancount['number_of_loans']-loancount['defsum']
loancount
```

```python
conditions = [
    (loancount['defaultdif']==0),
    (loancount['number_of_loans']==loancount['defaultdif']),
    (loancount['number_of_loans']>loancount['defaultdif'])
]

values = ['1','0','1']
loancount['prev_def'] = np.select(conditions,values)
loancount
```

```python
loancount.drop(['defsum','defaultdif'],inplace=True,axis=1)
loancount
```

```python
loancount = pd.merge(loancount,bal, on = 'UID')
loancount
```
